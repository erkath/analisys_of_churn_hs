```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import retentioneering
```

```python
data = pd.read_csv(r'C:\Users\Katerina\Питонячье\Diploma\modified_data.csv')
data
```

```python
data.columns
```

```python
data1 = data[['last_login', 'user_hash']].iloc[0:3000]
data1 = data1.rename(columns = {'last_login':'event_time_col'})
data1['event_col'] = 'last_login'
```

```python
data2 = data[['date_joined', 'user_hash']].iloc[0:3000] 
data2 = data2.rename(columns = {'date_joined':'event_time_col'})
data2['event_col'] = 'date_joined'
```

```python
data3 = data[['date_registered', 'user_hash']].iloc[0:3000]
data3 = data3.rename(columns = {'date_registered':'event_time_col'})
data3['event_col'] = 'date_registered'
```

```python
data_result = pd.concat([data1, data2, data3])
data_result
```

```python
retentioneering.config.update({
    'event_col':'event_col',
    'user_col': 'user_hash',
    'event_time_col':'event_time_col',
})
```

```python
data_result.rete.plot_graph(norm_type=None,
                     weight_col=None,
                     thresh=250)
```

```python
data_result.rete.get_clusters(method='kmeans',
                       n_clusters=8,
                       feature_type='tfidf',
                       ngram_range=(1,2));

data_result.rete.project(plot_type ='clusters',
#                   targets = ['cart'],
                  method='tsne');
```
