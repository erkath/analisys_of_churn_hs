```python
def date_to_datetime(data: pd.DataFrame) -> pd.DataFrame:
    for column in data.columns:
        print(column)
        try:
            data[column] = data[column].apply(datetime.fromisoformat)
            print('changed to datetime')
        except:
            print('not a datetime')
    
```

```python
DATE_LOAD = pd.Timestamp('2021-11-18 09:00:00')
```

```python
data['delta_registered_joined'] = (data.date_registered - data.date_joined).dt.total_seconds() 
data
```

```python
dummies_most_active_track = pd.get_dummies(data.the_most_active_track)
dummies_most_active_track = dummies_most_active_track.rename(
    lambda string: "most_active_track_" + string.replace(' ', '_'), axis='columns'
)
dummies_most_active_track
```

```python
data = pd.merge(data, dummies_most_active_track, left_index=True, right_index=True)
data
```

```python
SECONDS_IN_DAY = 3600 * 24
SECONDS_IN_WEEK = SECONDS_IN_DAY * 7
SECONDS_IN_MONTH = SECONDS_IN_DAY * 30
SECONDS_IN_YEAR = SECONDS_IN_DAY * 365

plt.figure()
plt.hist(data.time_since_last_login, bins=200, alpha=0.9)
plt.scatter([SECONDS_IN_DAY, SECONDS_IN_WEEK, SECONDS_IN_MONTH], [0,0,0],
           color='red', s=30)
plt.xlabel('time (sec)')
plt.ylabel('users')
plt.axvline(x=3 * SECONDS_IN_MONTH, color="orange", label='3 months')
plt.axvline(x=SECONDS_IN_YEAR, color="red", label='1 year')
plt.legend(loc='best')
plt.title('Время, прошедшее с последнего входа (time_since_last_login)')

plt.show()
print(SECONDS_IN_YEAR / 1e8)
```
